```python
import sqlite3
import pandas as pd
conn=sqlite3.connect('database.db')
cursor=conn.cursor()
#----create table------
cursor.execute(""" CREATE TABLE customers (cid INTEGER PRIMARY KEY ,order_name TEXT,amount INTEGER);""")
conn.commit()
data=pd.DataFrame({"cid":[1,2,3],"order_name":["egg","orange","apple"],"amount":[123,234,345]})
data.to_sql("customers", conn, if_exists="replace", index=False)
df=pd.read_sql('SELECT * FROM customers', conn)
print(df)
```
